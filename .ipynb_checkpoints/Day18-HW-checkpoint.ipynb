**Exercise 1** Logistic Regression with ONE explanatory variable.

Choose another variable from the data set to use as your explanatory variable and create a Logistic Regression model to predict if an email is spam or not. You should do all of the following:

I am selecting 'exclaim_mess' as my predictor for whether or not an email is spam or not because it quantifies urgency by counting the number of exlamation points in an email. It's typical for scammers to use that urgent sense of emotion to coerce victims into providing sensitive information and/or money. I am also choosing this variable to apply the log function because, as you can see above, it follows an exponential trend and the log function will normalized its behavior. 

```python
df_model = df[['exclaim_mess','spam']]

x = df_model['exclaim_mess'].values.reshape(-1,1)
y = df_model['spam']

lm = LogisticRegression()
lm.fit(x,y)

print('Classes:')
print(lm.classes_)
print('Coefficients:')
print(lm.coef_)
print('Intercept:')
print(lm.intercept_)
```

```python
df_model['prediction'] = lm.predict(x)

y_pred = df_model['prediction'].values

cnf_matrix = metrics.confusion_matrix(y, y_pred)
cnf_matrix
```

```python
TP = cnf_matrix[0,0] # The email was not spam and was correctly labeled 0
FP = cnf_matrix[0,1] # The email was spam and was incorrectly labeled as 0
FN = cnf_matrix[1,0] # The email was not spam at was incorrectly labeled 1
TN = cnf_matrix[1,1] # The email was spam and was correctly labeled 1
```

```python
# Create new columns in your data frame
df_model[['prob not spam','prob spam']] = lm.predict_proba(x)
```

```python
df_model.head()
```

I believe the cutoff point should be about 10%. If this this similar to an significance level for a hypothesis test, then there is a 10% chance of a false positive. We do not want to mark non-spam emails as spam. It could cost sensitve emails being ignored. 

```python
# Choose your cutoff
cutoff = .1 #if prob spam is 5% or higher it is labeled spam

# Use a lambda on the new columns to create your new prediction
df_model['new_prediction'] = df_model['prob spam'].apply(lambda x: 1 if x>cutoff else 0)

# Look at the model DF
df_model.head()
```

```python
# Redo the confusion matrix

y_pred = df_model['new_prediction'].values

cnf_matrix = metrics.confusion_matrix(y, y_pred)
cnf_matrix
```

```python
TP = cnf_matrix[0,0]
FP = cnf_matrix[0,1]
FN = cnf_matrix[1,0]
TN = cnf_matrix[1,1]

print('False Negative Rate:')
print(FN/ (TP+FN))
print('------')
print('False Positive Rate:')
print(FP/ (FP+TN))
print('------')
print('Sensitivity:')
print(1 - (FN/ (TP+FN)))
print('------')
print('Specificity:')
print(1 - (FP/ (FP+TN)))
```

False negative means an email not spam and it was labeled spam. This happened in 9.3% of cases where the email was not spam.

False positive means that am email was spam and was labeled not spam. This occured in 77% of cases. 

If the email was not spam it has a 90.7% probability of being labeled not spam.

If the email was spam it has a 22.22% probability of being labeled spam.
